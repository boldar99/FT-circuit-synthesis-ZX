```python
from fault_equivalent_rewrites import unfuse_n_3FE

import pyzx as zx
from pyzx import VertexType, EdgeType, Graph
from fault_equivalent_rewrites import *

N = 17

g = Graph()
v = g.add_vertex(VertexType.Z, N//2, 0)
for i in range(N):
    nv = g.add_vertex(VertexType.BOUNDARY, 2 * (i // 2) - 4, (2 * (i % 2) - 1) * 5)
    g.add_edge((v, nv))

unfuse_n_4FE(g, v)
zx.draw(g)
```
