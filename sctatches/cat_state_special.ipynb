```python
from spidercat.graphs_circular import _find_valid_l
import networkx as nx

from spidercat.draw import draw_spanning_forest_solution
from spidercat.spanning_tree import find_min_height_roots, match_forest_leaves_to_marked_edges
from spidercat.generate import cat_state_FT_spectial

N = 12
G, S, M = cat_state_FT_spectial(N)
print(_find_valid_l(N))

matchings = match_forest_leaves_to_marked_edges(S, M)
roots = find_min_height_roots(S)

draw_spanning_forest_solution(G, S, M, matchings, roots, figsize=(6, 6), pos=nx.circular_layout(G))
```

```python
from spidercat.markings import find_marking_property_violation

print(f"Verifying marking property...")
violations = find_marking_property_violation(G, M, N)
print(f"Verification result:\n", violations)
```
