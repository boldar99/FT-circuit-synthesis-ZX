```python
from spidercat.draw import draw_forest_on_graph
import networkx as nx

def G_F_alt_for_t_1(N):
    G = nx.Graph()
    G.add_nodes_from([0])
    G.add_nodes_from(range(2, 2 + N), is_mark=True)
    G.add_edge(0, 2)
    G.add_edge(0, 3)
    for i in range(N - 2):
        G.add_edge(2 + i, 4 + i)
    G.add_edge(N, N + 1)
    F = G.copy()
    F.remove_edge(N + 1, N)
    return G, F

G, F = G_F_alt_for_t_1(6)
draw_forest_on_graph(G, F)
```

```python
from spidercat.circuit_extraction import CatStateExtractor, StimBuilder

extractor = CatStateExtractor(StimBuilder(), verbose=True)
circ = extractor.extract(G, F, {0: 0})
circ.diagram('timeline-svg')
```

```python
def G_F_alt_for_t_0(N):
    G = nx.Graph()
    G.add_nodes_from([0])
    G.add_nodes_from(range(1, N + 1), is_mark=True)
    for i in range(N):
        G.add_edge(i, i + 1)
    F = G.copy()
    return G, F

G, F = G_F_alt_for_t_0(4)
draw_forest_on_graph(G, F)

extractor = CatStateExtractor(StimBuilder(), verbose=True)
circ = extractor.extract(G, F, {0: 0})
circ.diagram('timeline-svg')
```
